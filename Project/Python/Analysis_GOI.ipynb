```python
plt.title("Category",fontsize=20)
d['Category'].value_counts().plot.pie(autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
d.groupby('Country/State/Territories')[['Literacy_2001','Literacy_2011',]].sum().plot.bar(color=['green','orange'],
                                                                                          figsize=(8,5))
plt.title("Literacy Rate of different Countries",fontsize=20,color="red")
plt.ylabel('Category')
plt.show()
```

```python
d.groupby('Country/State/Territories')[['Literacy_Rural_2001','Literacy_Rural_2011',]].sum().plot.bar(color=
                                                                  ['green','orange'],figsize=(8,5))
plt.title("Literacy Rate of Rural area in different Countries",fontsize=20,color="red")
plt.ylabel('Category')
plt.show()
```

```python
d.groupby('Country/State/Territories')[['Literacy_2001','Literacy_2011','Literacy_Rural_2001','Literacy_Rural_2011']].sum().plot.bar(color
                                                                    =['red','green','yellow','pink'],figsize=(15,10))
plt.title("Country/State/Territories",color="Green",fontsize=25)
plt.ylabel('Category')
plt.show()
```
