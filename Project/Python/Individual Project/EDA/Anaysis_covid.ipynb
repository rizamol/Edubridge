# Grouping different type of cases as per the date  

```python
datewise= d.groupby(['Date']).agg({'Confirmed':'sum','Cured':'sum','Deaths':'sum'})
datewise
```

# To plot Confirmed cases in Kerala

```python
d_kerala=d[d['State']=='Kerala']
d_kerala
```

```python
plt.bar(d_kerala['Year'],d_kerala["Deaths"],color="red")
plt.title("Deaths Cases in Kerala",fontsize=15,color='green')
plt.xlabel('Confirmed',fontsize=15)
plt.ylabel('Year',fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.barh(d['Year'],d["Confirmed"],color="red")
plt.title("Confirmed Cases in Kerala",fontsize=15,color='green')
plt.xlabel('Confirmed',fontsize=15)
plt.ylabel('Year',fontsize=15)
plt.grid(True)
plt.show()
```

# Confirmed cases in India

```python
plt.bar(d['Year'],d["Confirmed"],color="red")
plt.title("Confirmed Cases in India",fontsize=15,color='green')
plt.xlabel('Confirmed',fontsize=15)
plt.ylabel('Year',fontsize=15)
plt.grid(True)
plt.show()
```

```python
d.groupby("State")[["Confirmed","Deaths","Cured"]].sum().plot.bar(color=["red","black","green"],figsize=(12,7))
plt.title("Covid cases",fontsize=15,color='green')

plt.ylabel('State',fontsize=15)
plt.grid(True)
plt.show()


```
