# Data Visualization

# Total Medal Basis of year

```python
plt.bar(d['Year'],d['Total Medals'])
plt.title('Year Vs Total Medals',fontsize=15,color='red')
plt.xlabel('Year',fontsize=15,color='green')
plt.ylabel('Total Medals',fontsize=15,color='green')
plt.grid(True)
plt.show()
```

# Medals Range in different years

```python
d.groupby('Year')[['Gold Medals','Silver Medals','Bronze Medals']].sum().plot.bar(color=['gold','silver','brown']
                                                                                          ,figsize=(15,10))
plt.title("Year wise different Medals Range",color="Green",fontsize=25)
plt.ylabel('Total Medals')
plt.show()
```

# To plot Medals Category Of India

```python
d_India.groupby('Year')[['Gold Medals','Silver Medals','Bronze Medals']].sum().plot.bar(color=['gold','silver','brown']
                                                                                        ,figsize=(15,10))
plt.title("Total Medals in India",Color="green",fontsize=25)
plt.ylabel('Total Medals')
plt.show()
```
