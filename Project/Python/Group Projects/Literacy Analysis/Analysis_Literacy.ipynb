# Data Visualization

# To Plot literacy rate of Urban area in 2011

```python
plt.bar(d['Literacy_Total_2011'],d['Literacy_Urban_2011'])
plt.title('Literacy of Urban area in 2011',fontsize=25,color='red')
plt.xlabel('Literacy Total 2011',color='green',fontsize=15)
plt.ylabel('Literacy Urban 2011',color='green',fontsize=15)
plt.grid(True)
plt.show()
```

# To Plot literacy rate of Rural area in 2011

```python
plt.bar(d['Literacy_Total_2011'],d['Literacy_Rural_2011'])
plt.title('Literacy Rural in 2011',fontsize=25,color='red')
plt.xlabel('Literacy Total 2011',color='green',fontsize=15)
plt.ylabel('Literacy Rural 2011',color='green',fontsize=15)
plt.grid(True)
plt.show()
```

# To Plot literacy rate of Urban area in 2001

```python
plt.barh(d['Literacy_Total_2001'],d['Literacy_Urban_2001'],color="brown")
plt.title('Literacy rate of Urban in 2001',fontsize=25,color='gold')
plt.xlabel('Literacy Total 2001',color='green',fontsize=15)
plt.ylabel('Literacy Urban 2001',color='green',fontsize=15)
plt.grid(True)
plt.show()
```

# To Plot literacy rate of Rural area in 2001

```python
plt.barh(d['Literacy_Total_2001'],d['Literacy_Rural_2001'],color="green")
plt.title('Literacy of Rural in 2001',fontsize=25,color='red')
plt.xlabel('Literacy Total 2001',color='green',fontsize=15)
plt.ylabel('Literacy Rural 2001',color='green',fontsize=15)
plt.grid(True)
plt.show()
```

# To show category basis of literacy rate

```python
plt.title("Category",fontsize=20)
d['Category'].value_counts().plot.pie(autopct="%1.1f%%",shadow=True)
plt.show()
```

# To plot Total Litearacy rate(2001 & 2011) of differet countries

```python
d.groupby('Country/State/Territories')[['Literacy_Total_2001','Literacy_Total_2011',]].sum().plot.bar(color=['green','orange'],
                                                                                          figsize=(8,5))
plt.title("Literacy Rate of different Countries",fontsize=20,color="red")
plt.ylabel('Category')
plt.show()
```

# To show a rural litearcy rate in rural area(2001 & 2011)

```python
d.groupby('Country/State/Territories')[['Literacy_Rural_2001','Literacy_Rural_2011',]].sum().plot.bar(color=
                                                                  ['green','orange'],figsize=(8,5))
plt.title("Literacy Rate of Rural area in different Countries",fontsize=20,color="red")
plt.ylabel('Category')
plt.show()
```

# To show a Urban litearcy rate in rural area(2001 & 2011)

```python
d.groupby('Country/State/Territories')[['Literacy_Urban_2001','Literacy_Urban_2011',]].sum().plot.bar(color=
                                                                  ['green','orange'],figsize=(8,5))
plt.title("Literacy Rate of Urban area in different Countries",fontsize=20,color="red")
plt.ylabel('Category')
plt.show()
```
